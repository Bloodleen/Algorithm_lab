### Бинарное дерево:

```python
class Tree:
    class _Node:

        def __init__(self, rd=None):
            self.znach = None
            self.roditel = rd
            self.right = None
            self.left = None

        def is_left(self):
            return self.roditel.left == self

        def is_koren(self):
            return self.roditel == None

    def __init__(self):
        self._koren = Tree._Node()

    def __contains(self, inp, node=None):
        if node == None:
            node = self._koren
        if node.znach == None:
            return False, node
        if node.znach == inp:
            return True, node
        if (inp > node.znach):
            return self.__contains(inp, node.right)
        if (inp < node.znach):
            return self.__contains(inp, node.left)

    def contains(self, inp):
        c, _ = self.__contains(inp)
        return c

    def add(self, inp):
        c, n = self.__contains(inp)
        if not c:
            n.znach = inp
            n.right = Tree._Node(n)
            n.left = Tree._Node(n)
        return not c

    def __find(self, mode, node=None):
        if node == None:
            node = self._koren
        if mode:
            if node.left.znach != None:
                return self.__find(mode, node.left)
        else:
            if node.right.znach != None:
                return self.__find(mode, node.right)
        return node.znach, node

    def __remove_edge(self, node):
        node.left.roditel = None
        node.right.roditel = None
        node.left = None
        node.right = None
        node.znach = None

    def __remove(self, node):
        if node.left.znach == None:
            node.left.roditel == None
            if node.is_koren():
                node.right.roditel = None
                self._koren = node.right
            else:
                node.right.roditel = node.roditel
                if node.is_left():
                    node.roditel.left = node.right
                else:
                    node.roditel.right = node.right
                node.roditel = None
            node.right = None
            node.left = None
            node.znach = None
            return
        if node.right.znach == None:
            node.right.roditel == None
            if node.is_koren():
                node.left.roditel = None
                self._koren = node.left
            else:
                node.left.roditel = node.roditel
                if node.is_left():
                    node.roditel.left = node.left
                else:
                    node.roditel.right = node.left
                node.roditel = None
            node.right = None
            node.left = None
            node.znach = None
            return
        _, t = self.__find(True, node.right)
        t.left.roditel = None
        t.left = node.left
        node.left.roditel = t
        if node.is_koren():
            node.right.roditel = None
            self._koren = node.right
        else:
            node.right.roditel = node.roditel
            if node.is_left():
                node.roditel.left = node.right
            else:
                node.roditel.right = node.right
            node.roditel = None
        node.right = None
        node.left = None
        node.znach = None
        return

    def remove(self, inp):
        c, n = self.__contains(inp)
        if c:
            if (n.left.znach == None) and (n.right.znach == None):
                self.__remove_edge(n)
                return c
            self.__remove(n)
        return c
        
    def min(self):
        c, _ = self.__find(True)
        return c

    def max(self):
        c, _ = self.__find(False)
        return c
    
    def __to_list(self, arr, node):
        if node.left.znach != None:
            self.__to_list(arr, node.left)
        arr.append(node.znach)
        if node.right.znach != None:
            self.__to_list(arr, node.right)

    def tolist(self):
        ret = []
        self.__to_list(ret, self._koren)
        return ret
```

### Простое рехэширование

```python
class MHMProst:
   
    def __init__(self, size=256):
        self.__size = size
        self.__arr = [None] * size

    def __myhash(inp):
        mhash = 5 ** (ord(inp[0]) % 7)
        mhash = mhash * (13 ** (ord(inp[-1]) % 7))
        mhash = mhash * len(inp)
        return mhash

    def __key(self, inp):
        return MHMProst.__myhash(inp) % self.__size

    def add(self, inp):
        k = self.__key(inp)
        if self.__arr[k] == None:
            self.__arr[k] = inp
            return True
        ik = k
        k += 1
        k = k % self.__size
        while (self.__arr[k] != None) and (k != ik):
            k += 1
            k = k % self.__size
        if (self.__arr[k] == None):
            self.__arr[k] = inp
            return True
        return False

    def __contains(self, inp):
        k = self.__key(inp)
        if self.__arr[k] == inp:
            return True, k
        ik = k
        k += 1
        k = k % self.__size
        while (self.__arr[k] != inp) and (k != ik):
            k += 1
            k = k % self.__size
        if (self.__arr[k] == inp):
            return True, k
        return False, None

    def contains(self, inp):
        r, _ = self.__contains(inp)
        return r

    def remove(self, inp):
        c, k = self.__contains(inp)
        if c:
            self.__arr[k] = None
        return c

    def tolist(self):
        ret = []
        for i in self.__arr:
            if i != None:
                ret.append(i)
        return ret
        
```

```python
import random as r
ferzi = []
while len(ferzi) < 8:
    ferzi = []
    doskat = [[i, j] for i in range(1,9) for j in range(1,9)]
    for _ in range(8):
        t = r.randint(0,len(doskat)-1)
        ferzi.append(doskat.pop(t))
        rm = []
        for i, e in enumerate(doskat):
            if e[0] == ferzi[-1][0]:
                rm.append(i)
                continue
            if e[1] == ferzi[-1][1]:
                rm.append(i)
                continue
            if (e[0] + e[1]) == (ferzi[-1][0] + ferzi[-1][1]):
                rm.append(i)
                continue
            if (e[0] - e[1]) == (ferzi[-1][0] - ferzi[-1][1]):
                rm.append(i)
        k = 0
        for i in rm:
            doskat.pop(i - k)
            k += 1
        if len(doskat) == 0:
            break
print(ferzi)
```
